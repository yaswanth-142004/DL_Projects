```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk', 'the glass of juice', 'the cup of tea', 'I am a good boy ', 'I am a good developer', 'understan the meaning of words', 'your videos are good']    
```

```python
### Define the vocabulary size 
voc_size = 10000
```

```python
## One hot Representation 

one_hot_repr = [one_hot(words,voc_size)for words in sent]
print(one_hot_repr)
```

```python
## Word embedding representation 
from tensorflow.keras.layers import Embedding 

from tensorflow.keras.utils import pad_sequences

from tensorflow.keras.models import Sequential 

```

```python
import numpy as np 

sent_len = 8 

embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)

print(embedded_docs)

```

```python
## Feasture Representation 

dim = 10
```

```python
model = Sequential()

model.add(Embedding(voc_size,dim))
model.compile('adam','mse')
```

```python
model.predict(embedded_docs)
```
